# 01. Exploratory Data Analysis (EDA) on UPI Transactions
This notebook generates synthetic UPI transactions and conducts exploratory analysis on transaction distributions and fraud indicators.

```python
from src.data.synthetic_data_generator import SyntheticUPIDataGenerator
import pandas as pd
import numpy as np

# Step 1: Generate dataset
generator = SyntheticUPIDataGenerator(num_accounts=200, num_transactions=4000)
df = generator.generate_dataset()
print(f"Dataset shape: {df.shape}")
df.head()
```

```python
# Step 2: Fraud Distribution
print("Fraud Class Balance:")
print(df['is_fraud'].value_counts())
print(df['is_fraud'].value_counts(normalize=True) * 100)
```

```python
# Step 3: Transaction Amounts Analysis
print("Transaction Amounts Summary by Class:")
print(df.groupby('is_fraud')['amount'].describe())
```
