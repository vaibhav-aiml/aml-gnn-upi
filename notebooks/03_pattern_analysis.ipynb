# 03. Money Laundering Pattern Mining & Analysis
This notebook analyzes graph topology patterns: Smurfing, Layering, and Round-tripping.

```python
from src.data.synthetic_data_generator import SyntheticUPIDataGenerator
from src.detection.patterns import PatternDetector
import networkx as nx

generator = SyntheticUPIDataGenerator(num_accounts=150, num_transactions=3000)
df = generator.generate_dataset()

G = nx.DiGraph()
for _, row in df.iterrows():
    G.add_edge(row['from_account'], row['to_account'], amount=row['amount'])

detector = PatternDetector(G)
patterns = detector.get_all_patterns()

print("Detected Smurfing:", len(patterns.get('smurfing', [])))
print("Detected Layering:", len(patterns.get('layering', [])))
print("Detected Round-tripping:", len(patterns.get('round_tripping', [])))
```
